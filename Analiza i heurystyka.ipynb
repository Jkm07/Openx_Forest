# Heurystyka i wstępna analiza danych

**Import danych**

## Krótka analiza

Możemy zwrócić uwagę, że dla pewnych typów lasów charakterystyczne są pewne rodzaje gleb. Wykorzystamy tego dla naszej strategi heurystycznej.

# Heurystyka

**Strategia**: Dla danego rekordu sprawdzamy jego Solid_Type. Następnie dla danego Solid_Type wybieramy ten label, który dla niego najczęściej występuje.

Nasz strategie realizuje klasa **MyHerusic(x)**

```python
class MyHeuristic:
    def __init__(self, sums):
        self.sums = sums
    def predict(self, x):
        soil_types = x.iloc[:,14:54].idxmax(axis=1)
        return soil_types.apply(lambda soil_type: self.sums[soil_type].idxmax())
    
my_heuristic = MyHeuristic(data.iloc[:,14:55].groupby(['Cover_Type']).sum())
```

***
**Wykonanie predykcji na danych**

```python
x = data.drop(['Cover_Type'], axis = 1)
y_pred = my_heuristic.predict(x)
y = data['Cover_Type']
```

***
**Porównanie wyników z oczekiwaniami**

```python
import sklearn.metrics as metrics
metrics.accuracy_score(y, y_pred)
```

```python
precision_scores = metrics.precision_score(y, y_pred, average=None)
recall_scores = metrics.recall_score(y, y_pred, average=None)
f1_scores = metrics.f1_score(y, y_pred, average=None)
```

```python
for i, label in enumerate(range(1, 8)):
    print("Label: %s" % str(label))
    print("Precision: %.3f" % precision_scores[i])
    print("Recall: %.3f" % recall_scores[i])
    print("F1: %.3f" % f1_scores[i])
```

## Podsumowanie

Nasza strategia dała ogólny wynik dokładności - 61%. Trzeba jednak pamiętać, że nasz model całkowicie ignoruje labele 4 i 5, przez co nie ma praktycznego zastosowania. Spowodowane jest to tym, że labeli 4 i 5 jest zdecydowanie mniej niż innych labeli i nie są one nigdzie najliczniejszą grupą jeśli chodzi o grupowanie rekordów na typy gleby.
